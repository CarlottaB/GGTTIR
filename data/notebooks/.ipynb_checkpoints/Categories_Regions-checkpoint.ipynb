- Ersetzen ist in diesem Fall keine gute Idee. Wir brauchen die Alten Kategorien noch um sie später mit dem anderen Dataset verbinden zu könne.


## Create Column for new clean categories:

```python
#function:
#die funktion sucht in einem Dictionary nach den keys und gibt den jeweiligen wert zurück

def cleaned_categories(old_category,dict_new_categories):
    for key,value in dict_new_categories.items():
        if old_category == key:
            return value
```
