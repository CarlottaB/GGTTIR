## Short notebook: combining dataframes on categories

```python
#import libraries
import pandas as pd
import re
import xlsxwriter
```

```python
#set file-paths
file_path_categories = '../data_base/Categories_over_time.xlsx'

file_path_thinktanks = '../data_base/CLEAN_THINKTANKS.xlsx'
```

```python
#import category-datasets
regions_category_df = pd.read_excel(file_path_categories,sheet_name='regions')
research_category_df = pd.read_excel(file_path_categories,sheet_name='research_areas')
special_category_df = pd.read_excel(file_path_categories,sheet_name='special_categories')
```

```python
#import think-tank datasets
regions_tt_df = pd.read_excel(file_path_thinktanks,sheet_name='regions')
research_tt_df = pd.read_excel(file_path_thinktanks,sheet_name='research_areas')
special_tt_df = pd.read_excel(file_path_thinktanks,sheet_name='special_categories')
```

```python
research_category_df['Category'] = research_category_df['Category'].str.lower().str.strip()
research_tt_df['Category'] = research_tt_df['Category'].str.lower().str.strip()
```

```python
research_tt_df.head()
```

```python
#merge datasets
cleand_regions_tt = pd.merge(regions_tt_df,regions_category_df[['Category','Category2']],left_on='Category',right_on='Category',how='left')

cleand_research_tt = pd.merge(research_tt_df,research_category_df[['Category','Category2']],left_on='Category',right_on='Category',how='left')

cleand_special_tt = pd.merge(special_tt_df,special_category_df[['Category','Category2']],left_on='Category',right_on='Category',how='left')
```

```python
#rename cateories:
cleand_regions_tt.rename(columns={'Category2':'cleaned_category'},inplace=True)

cleand_research_tt.rename(columns={'Category2':'cleaned_category'},inplace=True)

cleand_special_tt.rename(columns={'Category2':'cleaned_category'},inplace=True)
```
